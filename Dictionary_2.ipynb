```python
# 1. Inventory Tracker
# Scenario: Build an inventory tracker for a store that can help manage stock levels. The system should be able to:

# Add new items and their quantities to the inventory.
# Update the quantity of an existing item.
# Display a list of items with their current stock levels.
# Remove an item from the inventory.

# Answer me

# What data structures (lists/dictionaries) they would use and why.
# How they would handle user input and potential errors.
# Any additional features they think would improve the application.
inventory=[]
while True:
  print("\nInventory Tracker")
  print("1. Add new items")
  print("2. Update the quantity")
  print("3. Display all items")
  print("4. Remove Item")
  print("5. End the session")

  choice=input("\nChoose an option(1-5): ")

  if choice=="1":
    name=input("Enter the name of the item: ")
    quantity=int(input(f"Enter the quantity for {name}: "))
    item={"name":name,"quantity":quantity}
    inventory.append(item)
    print("Item has been added")

  elif choice=='2':
    name=input("Enter the name of the item: ")
    found=False
    for item in inventory:
      if item['name']==name:
        new_quantity=int(input("Enter the updated quantity: "))
        item["quantity"]=new_quantity
        print(f"Quntity for {item['name']} has been updated.")
        found=True
        break

      if not found:
        print(f"Item {name} is not found.")

  elif choice=='3':
    if not inventory:
      print("Inventory is empty")
    else:
      print('Item\t:\tQuantity')
      for item in inventory:
        print(f"{item['name']} : {item['quantity']}")

  elif choice=="4":
    name=input("Enter the name of item to remove: ")
    for item in inventory:
      if item['name']==name:
        # del inventory[item]             #del function takes index
        inventory.remove(item)
        print(f"{name} has been removed")
      else:
        print(f"{item['name']} doesn't exist in the list")

  elif choice=='5':
    print("Session has ended.")
    break

  else:
    print("Invalid choice. Please enter valid option.")
```

```python
# Student Management System

students = [] # empty list

while True:
  print("Student Management System")
  print("1. Add Student")
  print("2. Display Students")
  print("3. Update Student Grade")
  print("4. Delete Student")
  print("5. Exit")

  choice = input("Choose an option (1-5):")

  if choice == '1':
    name = input("Enter student's name: ")
    grade = input("Enter student's grade: ")
    student = {'name': name, 'grade': grade}
    students.append(student)
    print(f"Student {name} has been added successfully!\n") #\n this means line break

  elif choice == '2':
    if not students:
      print("No students found.\n")
    else:
      print("Student Records: ")
      for student in students:
        print(f"Name: {student['name']}, Grade: {student['grade']}")
      print() # for better formatting

  elif choice == '3':
      name = input("Enter the name of the student whose grade you want to update: ")
      found = False # This variable is used to track whether the student with the specified
      # name exists in the list of students.
      for student in students:
        if student['name'] == name:
          new_grade = input("Enter the new grade:")
          student['grade'] = new_grade
          print(f"Grade for {name} has been updated to {new_grade}.\n")
          found = True # This will indicates that the student was found and their grade was updated.
          break

      if not found:
        print(f"Student {name} not f\]ound.\n")

  elif choice=='5':
    print("Session has ended.\nHave a great day")
    break

  else:
    print("Invalid choice. Please enter a valid option.\n")
```
